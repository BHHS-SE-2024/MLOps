### Model Training - Polynomial Regression

We will next train a polynomial regression machine learning model. You will discover that this data set does not lend itself to polynominal regression

#### Step 1

Load the required dependencies:

```python
# Import frameworks
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('../../style_Matplotlib_charts.mplstyle')
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

#### Step 2

Store the data columns for Linear Regression.

```python
# Create a data frame from the csv file
training_data = pd.read_csv('2.3.1.training_data.csv')

x_name = 'TC'
y_name = 'Target'
x = np.array(training_data[x_name]).reshape(-1, 1)
y = np.array(training_data[y_name])
```

#### Step 3

Then save an instance of PolynomialFeatures, adjust the 'degree' parameter to suit your model and create the feature using teh data in 'x'.

```python
poly = PolynomialFeatures(degree=3, include_bias=False)
poly_features = poly.fit_transform(x)
```

#### Step 4

Plot the feature `x` and target `y` data on a graph using the column headings as the graph labels.

```python
# Plot the data points
plt.scatter(x, y, marker='x', c='r')
# Set the title
plt.title("Diabetes Disease Progress")
# Set the y-axis label
plt.ylabel(f'Training {training_data[x_name].name}')
# Set the x-axis label
plt.xlabel(f'Training {training_data[y_name].name}')
plt.show()
```

#### Step 5

Use the [Scikit-learn](https://scikit-learn.org/stable/) library to instantiate a Machine Learning model class and store it in the `my_model` variable as an object.

```python
# Create the model
my_model = LinearRegression()
# Fit the model to the data
my_model.fit(poly_features, y)
```

#### Step 7

Plot the features, targets and model (linear regression).

```python
y_pred = my_model.predict(poly_features)
plt.plot(x, y_pred)
plt.scatter(x, y, marker='x', c='r')
plt.title("Diabetes Disease Progress")
plt.ylabel(f'Training {training_data[x_name].name}')
plt.xlabel(f'Training {training_data[y_name].name}')
plt.show()
```
